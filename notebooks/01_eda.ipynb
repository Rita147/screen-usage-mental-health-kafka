```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8-whitegrid")
```

```python
raw_path = "../data/raw/screen_time_mental_health.csv"
df_raw = pd.read_csv(raw_path)

df_raw.head()
```

```python
df_raw.info()
```

```python
df_raw.describe()
```

```python
df_raw.isna().sum()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_raw["daily_screen_time_hours"], kde=True)
plt.title("Daily Screen Time Distribution")
plt.xlabel("Hours")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_raw["stress_level"], bins=10, kde=False)
plt.title("Stress Level Distribution")
plt.xlabel("Stress Level")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df_raw,
    x="daily_screen_time_hours",
    y="stress_level",
    alpha=0.6
)
plt.title("Screen Time vs Stress Level")
plt.xlabel("Daily Screen Time (hours)")
plt.ylabel("Stress Level")
plt.show()
```

```python
cols_of_interest = [
    "daily_screen_time_hours",
    "sleep_duration_hours",
    "weekly_anxiety_score",
    "weekly_depression_score",
    "stress_level",
    "mood_rating",
    "mental_health_score"
]

plt.figure(figsize=(10,8))
sns.heatmap(df_raw[cols_of_interest].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df_raw, x="gender", y="daily_screen_time_hours")
plt.title("Screen Time by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df_raw, x="location_type", y="mental_health_score")
plt.title("Mental Health Score by Location Type")
plt.show()
```
